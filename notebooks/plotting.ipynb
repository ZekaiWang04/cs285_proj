```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("other_experiments/trained_on_random_constant_0.05_short_losses", "rb") as f:
    losses_20 = pickle.load(f)
with open("other_experiments/trained_on_random_constant_0.05_40_losses", "rb") as f:
    losses_40 = pickle.load(f)
with open("other_experiments/trained_on_random_constant_0.05_medium_losses", "rb") as f:
    losses_80 = pickle.load(f)
with open("other_experiments/trained_on_random_constant_0.05_long_losses", "rb") as f:
    losses_200 = pickle.load(f)
fig, ax = plt.subplots(figsize=(8, 6))
t = np.arange(1000)
ax.plot(t, losses_20, label="ep_len = 20")
ax.plot(t, losses_40, label="ep_len = 40")
ax.plot(t, losses_80, label="ep_len = 80")
ax.plot(t, losses_200, label="ep_len = 200")
ax.legend()
ax.set_xlabel("Training Steps")
ax.set_ylabel("Prediction MSE Loss")
ax.set_title("Training with Different Episode Lengths")
plt.savefig("Training with Different Episode Lengths.png")
```

```python
steps = 100 * np.arange(1, 11)
on_policy_mean = np.array([-1163.0746, -1181.1393, -1143.2533, -1214.3406, -1198.2705, -1190.3972, -1270.0051, -1282.9028, -1189.7778, -1193.6011])
on_policy_std = np.array([188.10399, 224.62195, 262.16217, 239.65506, 276.0532, 275.91638, 286.14655, 279.94925, 281.8926, 337.32697])
off_policy_mean = np.array([-1006.4391, -310.3934, -703.7748, -349.80783, -295.05695, -308.28378, -163.30235, -265.8725, -255.81967, -272.42526])
off_policy_std = np.array([248.86777, 248.86777, 537.1487, 193.91254, 155.57991, 168.95016, 133.69273, 195.94273, 179.65504, 214.60643])
fig, ax = plt.subplots(figsize=(8, 6))
ax.fill_between(steps, on_policy_mean - on_policy_std, on_policy_mean + on_policy_std, label="on_policy", alpha=0.5)
ax.fill_between(steps, off_policy_mean - off_policy_std, off_policy_mean + off_policy_std, label="off_policy", alpha=0.5)
ax.plot(steps, on_policy_mean)
ax.plot(steps, off_policy_mean)
ax.legend()
ax.set_xlabel("Training Steps")
ax.set_ylabel("Average Return")
ax.set_title("On-Policy vs Off-Policy")
plt.savefig("On-Policy vs Off-Policy.png")
```
